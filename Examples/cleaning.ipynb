# Example of cleaning artifacts from a signal

```python
PATH = '' # Where the repository was cloned
import sys
sys.path.append(PATH)

from Code.utilities import read_signal, save_cleaned_signal
from Code.visualization import plot_signal
from Code.signal_cleaning import clean_signal, remove_artifacts
```

```python
SIGNAL_PATH = '' # Path to .csv file
signal = read_signal(SIGNAL_PATH, signal_name='fvr', sep=';')
```

```python
plot_signal(signal['signal'])
```

```python
plot_signal(signal['signal'], start_time=270000, end_time=280000)
```

```python
cleaned_signal = clean_signal(signal['signal'], window_size=9,
                      threshold_multiplier=3, interpolation_method='slinear', 
                 additional_interpolation_method='slinear')

cleaned_signal = clean_signal(cleaned_signal['cleaned'], window_size=7,
                      threshold_multiplier=4, interpolation_method='previous', 
                 additional_interpolation_method=None)
```

```python
plot_signal(cleaned_signal['cleaned'], start_time=270000, end_time=280000)
```

```python
cleaned_signal = remove_artifacts(cleaned_signal['cleaned'], 30)
```

```python
save_cleaned_signal(SIGNAL_PATH, signal_name='fvr', cleaned_signal=cleaned_signal)
```
